## 1차 데이터 전처리

##### 1. 필요없는 문자 제거 (~~호)

```python
df_col = []
for i in range(len(df['세부내용'])):
    romove_col = df['세부내용'][i].replace('호',"")
    df_col.append(romove_col)
df['세부내용'] = df_col
df['세부내용']
```

##### 2. 특수 문자 제거

```python
df['세부내용'] = df['세부내용'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
```

```python
df['세부내용']
```

### 3. 불용어 처리

```python
sentences = []

stopwords = ['고객요청사항', '진단결과', '회선단말특이사항', '기타추가연락처']

for sentence in df['세부내용']:
    for word in stopwords:
        sentence = sentence.replace(word, '')
        sentence = sentence.lstrip()
        sentence = sentence.replace(' ', '')
    sentences.append(sentence)
df['sen'] = sentences
```

```python
df['sen']
```

##### 3. 자동 띄어 쓰기 적용

```python
from pykospacing.kospacing import *

spacing = Spacing()
```

```python
sample_data  = df['sen'][0]
kospacing_sent = spacing(sample_data) 

print('자동 띄어 쓰기 적용 전 :', sample_data)
print('자동 띄어 쓰기 적용 후 :', kospacing_sent)
```

```python
## 띄어쓰기 로직을 데이터 셋에 적용


clean_data = []
for row in df.iterrows() :
    start_time = timeit.default_timer()
    
    spacings = spacing(row[1]['sen'])
    clean_data.append(spacings)

    terminate_time = timeit.default_timer() # 종료 시간 체크  
print("%f초 걸렸습니다." % (terminate_time - start_time))
clean_data
```

```python
preprocess_data = pd.DataFrame({'Label':df['VOC 유형'],
                             'sentence':clean_data})
```

```python
preprocess_data.to_csv('../dataset/CAI_data_001.csv', index=False)
```
