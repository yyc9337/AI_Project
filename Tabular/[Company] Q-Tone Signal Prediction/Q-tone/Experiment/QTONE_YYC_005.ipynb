```python
import pandas as pd
```

```python
val_df = pd.read_csv('../Common/Validation_dataframe/Validation_dataframe.csv')
test_df = pd.read_csv('../Processing/data/Type_A/AXN.csv')
```

```python
test = test_df[:1]
```

```python
from sys import displayhook


displayhook(val_df.head())
displayhook(test.head())
```

```python
def case2_validation(df, test) :
    for i in df['channel'] :
        if i == test['채널명'][0] :
            #print(df[df['channel'] == test['채널명'][0]])

            val_frame = df.loc[df['channel'] == i ]
            #print(val_frame['input_max'][66])
            
            if val_frame['input_max'][66] > test['삽입시간'][0]:
                print('정상')
            elif val_frame['input_max'][66] < test['삽입시간'][0]:
                print('비정상')
```

```python
case2_validation(val_df, test)
```
