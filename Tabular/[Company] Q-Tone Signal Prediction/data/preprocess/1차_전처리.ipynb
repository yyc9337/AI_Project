```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv('../Common/data/전체신호.csv', encoding = 'cp949')
```

```python
df.head()
```

```python
df['큐톤삽입시간'][0] = df['큐톤삽입시간'][0].replace('?','')
times = []
for i in range(len(df)) :
    time = df['큐톤삽입시간'][i][:19]
    times.append(time)
df['삽입시간'] = times
df = df.drop('큐톤삽입시간',axis=1)

df.head()
```

```python
df[df['광고타입'] == 'B'].index
```

## 전처리

```python
drop_index = df[df['광고타입'] == 'B'].index
drop_index2 = df[df['삽입시간'] == '?2021-10-16 00:00:2'].index
df = df.drop(drop_index,axis=0)
df = df.drop(drop_index2,axis=0)
df = df.reset_index()
data = df.drop(['index', '큐톤ID', 'OTV 채널번호', 'OTS 채널번호'],axis=1)
```

```python
data['날짜'] = data['삽입시간'].str.split(' ').str[0]
data['시간'] = data['삽입시간'].str.split(' ').str[1]
```

```python
for i in range(len(data)) :
    data['Date'][i] = data['날짜'][i].replace('-','')
    data['Date'][i] = data['날짜'][i].replace('?','')
    data['Time'][i] = data['시간'][i].replace(':','')
```

## 채널별 CSV 파일로 저장

```python
channel_name = df['채널명'].unique().tolist()
```

```python
for chat_name in channel_name :
    chat_csv = None
    chat_csv = df[df['채널명'] == chat_name]
    chat_csv.to_csv('../Processing/data/Type_A/'+str(chat_name)+'.csv', index=False)
```

## 전체 파일 CSV 저장

```python
data.to_csv('../Processing/data/QTONE_data_YYC_002.csv', index= False)
```
