```python
import os
import pandas as pd
import numpy as np

import matplotlib.pylab as plt
import seaborn as sns
%matplotlib inline

from sklearn.preprocessing import StandardScaler
from tensorflow.python.keras.models import load_model

import tensorflow as tf
from tensorflow import keras
```

```python
os.chdir('../../home/innerwave/ailab/채널AI/L3/Experiment/')
path = '../Processing/bps_out/'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')] ## 파일명 끝이 .csv인 경우
```

```python
file_list_py
```

```python
def make_vector(X, Y, time_steps=1):
    Xs, Ys = [], []
    for i in range(len(X) - time_steps):
        value = X.iloc[i:(i + time_steps)].values
        Xs.append(value)        
        Ys.append(Y.iloc[i + time_steps])
    return np.array(Xs), np.array(Ys)
```

```python
## 경로 재설정
os.chdir('../../home/innerwave/ailab/채널AI/L3/Experiment/')

for i in file_list_py:
    ## 데이터 초기화
    ticket = None

    # 데이터 로드
    df = pd.read_csv(path + i)
    ticket = pd.read_csv(path + i)
    time_temp = pd.to_datetime(ticket['stat_dt'])
    level_temp = ticket['bps_out']
    time_temp.name = 'date'
    level_temp.name = 'bps_out'

    ticket = pd.concat([time_temp, level_temp ], axis = 1)
    ticket = ticket.set_index("date")
    print(ticket.shape)
    ticket.head(5)

    plt.rcParams["figure.figsize"] = (20,10)
    ticket.plot()

    # Set Anomaly
    span_start = time_temp[4950].to_pydatetime()
    span_end = time_temp[5150].to_pydatetime()
    plt.axvspan(span_start, span_end, facecolor='r', alpha=0.5)

    plt.title("bps_out")
    plt.xlabel("Time")
    plt.ylabel("bps_out")
    plt.show()

    train_size = int(len(ticket) * 0.7)
    test_size = len(ticket) - train_size
    train, test = ticket.iloc[0:train_size], ticket.iloc[train_size:len(ticket)]
    print(train.shape, test.shape)

    from sklearn.preprocessing import StandardScaler

    scaler = StandardScaler()
    scaler = scaler.fit(train[['bps_out']])

    train['bps_out'] = scaler.transform(train[['bps_out']])
    test['bps_out'] = scaler.transform(test[['bps_out']])

    TIME_STEPS = 30

    X_train, Y_train = make_vector(train[['bps_out']], train.bps_out, TIME_STEPS)
    X_test, Y_test = make_vector(test[['bps_out']], test.bps_out, TIME_STEPS)

    model = keras.Sequential()
    model.add(keras.layers.LSTM(units=64, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(keras.layers.Dropout(rate=0.2))
    model.add(keras.layers.RepeatVector(n=X_train.shape[1]))
    model.add(keras.layers.LSTM(units=64, return_sequences=True))
    model.add(keras.layers.Dropout(rate=0.2))
    model.add(keras.layers.TimeDistributed(keras.layers.Dense(units=X_train.shape[2])))
    model.compile(loss='mae', optimizer='adam')

    history = model.fit(X_train, X_train, epochs=10, batch_size=32, validation_split=0.1, shuffle=False)
    
    X_pred = model.predict(X_train)
    train_mae_loss = np.mean(np.abs(X_pred - X_train), axis=1)

    sns.distplot(train_mae_loss, bins=50, kde=True);

    X_pred = model.predict(X_test)
    test_mae_loss = np.mean(np.abs(X_pred - X_test), axis=1)

    THRESHOLD = 0.3 # Manual Setting

    test_score_df = pd.DataFrame(index=test[TIME_STEPS:].index)
    test_score_df['loss'] = test_mae_loss
    test_score_df['threshold'] = THRESHOLD
    test_score_df['anomaly'] = test_score_df.loss > test_score_df.threshold
    test_score_df['bps_out'] = test[TIME_STEPS:].bps_out

    anomalies = test_score_df[test_score_df.anomaly == True]
    print(anomalies.head())


    model.save('../Model/bps_out/' +str(df['ip'][0])+'_bpsout.h5')

    ## MLflow Logging
    os.chdir('/mlflow-management/L3-mlflow/')
    
```
